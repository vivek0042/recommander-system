# popularity based recommander system

```python
bandhan=rating.merge(books,on="ISBN")
bandhan
```

```python
num_rating_df=bandhan.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={"Book-Rating":"num_ratings"},inplace=True)
```

```python
num_rating_df
```

```python
avg_rating_df=bandhan.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={"Book-Rating":"avg_ratings"},inplace=True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on="Book-Title")
```

```python
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings']>200].sort_values('avg_ratings',ascending=False).head(50)
```

```python
popular_df
```

```python
popular_df=popular_df.merge(books,on="Book-Title").drop_duplicates("Book-Title")[["Book-Title","Book-Author","num_ratings","avg_ratings","Image-URL-M"]]
```

## Collaborative Filtering Based approch

```python
x=bandhan.groupby('User-ID').count()['Book-Rating']>200
educated_users=x[x].index
# educated_users
```

```python
filtered_rating=bandhan[bandhan['User-ID'].isin(educated_users)]
# filtered_rating
```

```python
# now above two line code we use to find out number of educated users for that we define formula that 
# if users give rating on more than 200 books then he/she is educated and true rater and he has knowledge of rating.
```

```python
# now we go for the find how many total books which are  actually get nominial rating for that we have decide
# that we only considered that books which are get more than 50 ratings.
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
# famous_books=y[y].index
# famous_books
famous_books=y[y].index
famous_books
```

```python
final_rating=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_rating
```

```python
pt=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

```python
# in above table we have only those rows and  columns those ahve more than 50 ratings on books and  only those users 
# who gave more than 200 ratings.
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
# now we want to find cosine similarity with all other points.
# so here total 706 points 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores = cosine_similarity(pt)
similarity_scores
```

```python
similarity_scores.shape
```

```python
# we find cosine similarity just because to find out 5 nearest or we can say that highest voted books.
# through this we get that books
# here we make function to find out that books.
```

```python
def recommend(book_name):
    
    
#     first fetch index of book
    index = np.where(pt.index==book_name)[0][0]
#     next we find similar five books which are nearest to this book ranking
    similar_score = sorted(list(enumerate(similarity_scores[index])),key = lambda x:x[1],reverse = True) [1:6]  
    
    for i in similar_score:
        print(pt.index[i[0]])
```

```python
recommend('1984')
```
